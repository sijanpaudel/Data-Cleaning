## Removing Outliers using IQR

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan_data_set.csv")
dataset.head(3)
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="CoapplicantIncome", data=dataset)
plt.show()
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
```

```python
IQR = q3-q1
```

```python
min_range = q1 - (1.5*IQR)
max_range = q3 + (1.5*IQR)
min_range, max_range
```

```python
new_dataset = dataset[dataset["CoapplicantIncome"]<=max_range]
```

```python
new_dataset
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="CoapplicantIncome", data=new_dataset)
plt.show()
```

## Removing Outliers using Z-Score

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("loan_data_set.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.boxplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
min_range = dataset["CoapplicantIncome"].mean() - 3*dataset["CoapplicantIncome"].std()
max_range = dataset["CoapplicantIncome"].mean() + 3*dataset["CoapplicantIncome"].std()
min_range, max_range
```

```python
new_data = dataset[dataset["CoapplicantIncome"]<=max_range]
```

```python
new_data.shape
```

```python
sns.boxplot(y="CoapplicantIncome", data=new_data)
plt.show()
```

```python
z_score = (dataset["CoapplicantIncome"] - dataset["CoapplicantIncome"].mean()) / dataset["CoapplicantIncome"].std()
```

```python
dataset["z_score"] = z_score
```

```python
dataset[dataset["z_score"]<3]
```

# Function Transformer

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pd.read_csv("loan_data_set.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
iqr = q3-q1
```

```python
min_r = q1 - (1.5*iqr)
max_r = q3 + (1.5*iqr)
min_r, max_r
```

```python
dataset = dataset[dataset["CoapplicantIncome"]<=max_r]
```

```python
sns.distplot(dataset["CoapplicantIncome"])
plt.show()
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft = FunctionTransformer(func=np.log1p)
```

```python
ft.fit(dataset[["CoapplicantIncome"]])
```

```python
dataset["CoapplicantIncome_tf"] = ft.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_tf"])

plt.show()
```

```python
ft1 = FunctionTransformer(func=lambda x: x**2)
ft1.fit(dataset[["CoapplicantIncome"]])
dataset["CoapplicantIncome_tf1"] = ft1.transform(dataset[["CoapplicantIncome"]])
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["CoapplicantIncome"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["CoapplicantIncome_tf1"])

plt.show()
```
